```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#importing dataset and retreiving the independent and dependent variables
kyphosis = pd.read_csv("kyphosis.csv")

X = kyphosis.drop('Kyphosis',axis = 1)
y = kyphosis['Kyphosis']
```

```python
kyphosis.head(n=10)
```

```python
sns.barplot(x='Kyphosis',y='Age',data=kyphosis)
```

```python
sns.pairplot(kyphosis,hue='Kyphosis',palette='Set1')
```

```python
plt.figure(figsize = (25,7))
sns.countplot(x = 'Age', hue = 'Kyphosis', data = kyphosis, palette = 'Set1')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators = 100)

rfc.fit(X_train,y_train)
```

```python
pred = rfc.predict(X_test)
pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test,pred))
```

```python
#Evaluate the model by using confusiion matrix
cm = confusion_matrix(y_test, pred)
cm
```

```python
accuracy = (cm[0][0]+cm[1][1])/(sum(cm[0])+sum(cm[1]))
accuracy
```

```python
error = (cm[0][1]+cm[1][0])/(sum(cm[0])+sum(cm[1]))
error
```
